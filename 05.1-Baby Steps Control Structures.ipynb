```python
bmi=int(input("Enter the Your BMI Index :"))

if (bmi<16.5):
    print("Severely Underweight")
elif(bmi<18.5):
    print("Underweight")
elif(bmi<25):
    print("Normal")
elif(bmi<30):
    print("Overweight")
elif(bmi<35):
    print("Obese Class - 01")
elif(bmi<40):
    print("Obese Class - 02")
elif(bmi<45):
    print("Severely Obese")
elif(bmi<50):
    print("Morbidly Obese")
elif(bmi<60):
    print("Super Obese")
else:
    print("Hyper Obese")
```
